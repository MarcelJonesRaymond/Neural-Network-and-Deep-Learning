```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
import random
random.seed(0)
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([1, 3, 6, 5, 4, 7, 1, 6, 10, 13, 5])
y = np.array([3, 7, 13, 11, 9, 15, 3, 13, 21, 27, 11])
```

```python
model=Sequential()
model.add(Flatten())
model.add(Dense(1))
```

```python
model.compile(optimizer='SGD',loss='mse')
```

```python
model.fit(x,y,epochs=100)
```

```python
model.predict([100])
```

```python
from tensorflow.keras.datasets import boston_housing
(features,actual_prize),_=boston_housing.load_data(test_split=0)
```

```python
model=Sequential()
model.add(tf.keras.layers.BatchNormalization(input_shape=(13,)))
model.add(Dense(1))
```

```python
model.compile(optimizer='SGD',loss='mse')
```

```python
model.fit(features,actual_prize,epochs=500,validation_split=0.3)
```

```python
import numpy as np
test_x = np.reshape([1.2, 0, 8.14e+00, 0.0e+00, 5.3e-01, 6.14e+00, 9.170e+01, 3.97e+00, 4, 3.07e+02, 2.10e+01, 3.96e+02, 1.872e+01],(-1, 13))

test_y = model.predict(test_x)

print(test_y)
```
